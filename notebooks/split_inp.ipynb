```python
from pathlib import Path
```

```python
path = Path("../inp/Case_2nd_order_nlgeom.inp")
with path.open("r", encoding="utf-8") as f:
    lines = f.readlines()
```

```python
node_rows = []
part_rows = []
element_rows = []
for row, line in enumerate(lines):
    if "*NODE" in line:
        node_rows.append([row, line])
    if "Part" in line:
        part_rows.append([row, line])
    if "*ELEMENT" in line:
        element_rows.append([row, line])
```

```python
node_row_01 = node_rows[0][0]
element_row_01 = element_rows[0][0]
# test_lines = lines[node_row_01-10:element_row_01-4195500]
test_lines = lines[node_row_01-10:element_row_01-4000000]

path_w = Path("../inp_check_.inp")
with open(path_w, mode='w') as f:
    f.writelines(test_lines)
```
