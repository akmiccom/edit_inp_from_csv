```python
files = glob("../input/*.csv")
if len(files) > 1:
    print("csvファイルが複数あります。ファイルを一つにしてください。")
else:
    csv_path = files[0]
df_csv = pd.read_csv(csv_path, usecols=usecols_csv, dtype=dtypes)
df_csv = df_csv.rename(columns=renames)
check_size = 1000
df_csv.iloc[:check_size].to_csv(f"../input_check_{check_size}.csv", index=False)
df_csv.head()
```

```python
prospect_coefficient = 1
df_csv["X_new"] = df_csv["X"] + df_csv["U-U1"] * -1 * prospect_coefficient
df_csv["Y_new"] = df_csv["Y"] + df_csv["U-U2"] * -1 * prospect_coefficient
df_csv["Z_new"] = df_csv["Z"] + df_csv["U-U3"] * -1 * prospect_coefficient
df_csv.head()
```
